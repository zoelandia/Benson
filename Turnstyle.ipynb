```python
ts = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170304.txt')
```

```python
ts.head()
```

```python
ts['time_error'] = ts['TIME']
ts.head()
```

```python
ts.time_error = ts.time_error.str.split(':')
ts.head()
```

```python
ts['time_error'] = ts['time_error'].apply(lambda x: int(x[0]))
ts.head()
```

```python
ts['DATETIME'] = pd.to_datetime(ts['DATE'] + ' ' + ts['TIME'])
ts.head()
```

```python
ts['timet'] = ts['DATETIME'].apply(lambda L: pd.datetime(L.year, L.month, L.day, L.hour))
ts.head()
```

```python
ts['timet'] = ts['DATETIME'].values.astype('<M8[h]')
ts.head()
```
